```python
import pandas as pd
import pymongo


config = {
    "client": "mongodb://localhost:27017/",
    "database": "datascience_test"
}


# Define required functions


def mdb(collection: str):
    client = pymongo.MongoClient(config["client"])
    database = client[config["database"]]
    collection_ = database[collection]
    return collection_


def show(cursor):
    return pd.DataFrame(cursor)
```

```python
# Basisabfragen:
# 1. Zeige alle BÃ¼cher (Books) in der Datenbank an.
show(mdb("books").find())
```

```python
# Basisabfragen:
# 2. Finde alle Benutzer (Users) Ã¼ber 30 Jahre alt.
show(mdb("users").find({"Age": {"$gt": 30}}))
```

```python
# Basisabfragen:
# 3. Zeige die ersten 5 Bewertungen (Ratings) an.
show(mdb("ratings").find().limit(5))
```

```python
# Filtern und Sortieren:
# 1. Finde alle BÃ¼cher, die von einem bestimmten Autor (z.B. Jules Verne) geschrieben
# wurden
show(mdb("books").find({"Book-Author": "Jules Verne"}))
```

```python
# Filtern und Sortieren:
# 2. Zeige die BÃ¼cher an, die nach 2000 verÃ¶ffentlicht wurden, sortiert nach
# dem Jahr der VerÃ¶ffentlichung
show(mdb("books").find(
    {"Year-Of-Publication": {"$gt": 2000}}).sort("Year-Of-Publication"))
```

```python
# Filtern und Sortieren:
# 3. Finde die Top 5 Benutzer mit den meisten Bewertungen.
show(mdb("ratings").aggregate([{"$sortByCount": "$User-ID"}, {"$limit": 5}]))
```

```python
# Weitere Funktionen:
# 1. FÃ¼gen Sie einen neuen Benutzer in die "users"-Sammlung ein. Der neue Benutzer hat
# die User-ID 123, befindet sich in der Location "Berlin" und ist 25 Jahre alt.
mdb("users").insert_one({
    "User-ID": 123,
    "Location": "Berlin",
    "Age": 25
})
show(mdb("users").find({"User-ID": 123}))
```

```python
# Weitere Funktionen:
# 2. Aktualisieren Sie den Standort des Benutzers mit der User-ID 123 auf "MÃ¼nchen".
mdb("users").update_one({"User-ID": 123},
                        {"$set": {"Location": "MÃ¼nchen"}})
show(mdb("users").find({"User-ID": 123}))
```

```python
# Weitere Funktionen:
# 3. LÃ¶schen Sie alle BÃ¼cher in der "books"-Sammlung, die vor dem Jahr 2000
# verÃ¶ffentlicht wurden.
mdb("books").delete_many({"Year-Of-Publication": {"$lt": 2000}})
```

```python
# Weitere Funktionen:
# 4. Berechnen Sie den Durchschnitt des Alters aller Benutzer,
# die sich in der Stadt "nyc, new york, usa" befinden.
show(mdb("users").aggregate([
    {"$match": {
        "Location": "nyc, new york, usa"}},
    {"$group": {
        "_id": "null",
        "average_age": {"$avg": "$Age"}}}]))
```
